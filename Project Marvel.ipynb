```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Import File csv

```python
marvel= pd.read_csv("characters_stats.csv")
print(marvel.head())

marvel.shape
```

# HEROS

# Alignment

```python
# Categorical columnn, frequency counts:
# Nature of characters, Hero / Villian
marvel['Alignment'].value_counts()
```

```python
# Extracting good characters
good= marvel[marvel['Alignment']=='good']
print(good.shape)
good.head()
```

# Speed

```python
# Descending Sorintg with speed, by default ascending = True so we make it false.
good.sort_values('Speed', ascending= False).head()
```

# Power

```python
# Descending Sorintg with power, by default ascending = True so we make it false.
good.sort_values('Power', ascending= False).head()
```

```python
# For max power , power =100
mp= good[good['Power']==100]
print(mp.shape)
mp
```

# Best Super Hero 

```python
bsh= good.sort_values('Total', ascending=False)
bsh.head()
```

```python
plt.figure(figsize=(10,5))
# For bar plot (Names, Numerical Values)
# for 1st 5 characheters [:5] used:
plt.bar(bsh['Name'][:5],bsh['Total'][:5], color='orange')
plt.show()
```
