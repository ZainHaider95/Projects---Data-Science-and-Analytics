# Reading from matches.csv 

```python
ipl= pd.read_csv("matches.csv")
print(ipl.shape)
ipl.head()
```

# Most Man of the Matches

```python
# Selection from highest counts from Man of the matches:

ipl['player_of_match'].value_counts()
```

```python
# Selecting from 0 to 4
ipl_count5= ipl['player_of_match'].value_counts()[0:5]
ipl_count5
```

```python
# bar plot using more size:

plt.figure(figsize=(20,5))
plt.bar(ipl_count5_keys, ipl_count5, color='Brown' )
plt.show()
```

# Toss winners

```python

# Finding toss winner
ipl['toss_winner'].value_counts()
```

# Batting first winners:

```python
 # Winning by runs means batting first
bat_first= ipl[ipl['win_by_runs'] !=0]    
bat_first.head()
    
```

```python
# Showing only win by runs with index numbers
bat_first['win_by_runs']
```

```python
plt.figure(figsize=(15,7))

plt.hist(bat_first['win_by_runs'], color='Orange' )
plt.title('Winning by Runs : Distribution')
plt.xlabel('Runs')
plt.show()
```

```python
# Winners by Batting first:
bat_first['winner'].value_counts()
```

```python
# Winner of ipl by batting first top 3:

plt.figure(figsize=(10,5))

# plt.bar(x,y)
# .keys() is for giving names: Mumbai, Chennai, Punjab on X-axis
# stats of winners are in Y-axis 

plt.bar(list(bat_first['winner'].value_counts()[0:3].keys()) , bat_first['winner'].value_counts()[0:3])
plt.show()

#Improper Methord:
#plt.bar(['Mumbai','Chennai','Punjab'],bat_first['winner'].value_counts()[0:3])
```

```python
plt.figure(figsize=(15,15))

# plt.pie(strings, labels= numerical)
# for percentage autoPCT
plt.pie(bat_first['winner'].value_counts() ,labels= bat_first['winner'].value_counts().keys(), autopct='%0.01f%%')
plt.show()
```

# Reading from deliveries.csv 

```python
#Ball by ball information per season
deliveries= pd.read_csv("deliveries.csv")
print(deliveries.shape)
deliveries.head()
```

# Unique: Gives all number of matches

```python
print(deliveries.shape)
print("")

# Index used for bowling so for matches
deliveries['match_id'].unique()
```

# Match 1:

```python
# Match_id ==1 means match number 1:  SRH Vs RCB

m1= deliveries[deliveries['match_id']==1]

print(m1.head())
print("")
```

```python
# 248 records are the 8 extra balls (no ball, wide ball)
m1.shape
```

# SRH Batting 1st analysis:

```python
# SRH match 1 :

SRHm1= m1[m1['inning']==1]
SRHm1
```

```python
SRHm1['batsman_runs'].value_counts()

# 57 singles, 32 dot balls ......
```

```python
SRHm1['dismissal_kind'].value_counts()
```
