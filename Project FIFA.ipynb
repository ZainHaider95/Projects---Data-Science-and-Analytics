# Nationality

```python
fifa['nationality'].value_counts()
```

```python
# For first ten elements values

v=fifa['nationality'].value_counts()[:5]
v
```

```python
# For first ten elements keys

k= fifa['nationality'].value_counts()[:5].keys()
k
```

```python
# Bar plotting

plt.figure(figsize=(15,5))

#In Bar plot first paramaneter is STRING (catagorical) and second parameter is NUMBER (numerical):
plt.bar(k,v, color='pink')
plt.show()
```

# Salary

```python
salarly= fifa[['short_name','wage_eur']]
salarly.head()
```

```python
# Sorting salarly in same variable:
salarly= salarly.sort_values('wage_eur', ascending= False)
salarly.head()
```

```python
plt.figure(figsize=(15,5))
plt.bar(salarly['short_name'][:5],salarly['wage_eur'][:5],color=['y','b','g','r','brown'])
plt.show()
```
